```python
# Importing necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error,mean_absolute_error,median_absolute_error,r2_score
import tkinter as tk
from tkinter import messagebox
```

```python
# Read dataset 
df = pd.read_csv('NFLX.csv')
```

```python
# Split data into train and test sets
train, test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
# Define features and target variables
x_train = train[['Open', 'High', 'Low', 'Volume']].values
x_test = test[['Open', 'High', 'Low', 'Volume']].values
y_train = train['Close'].values
y_test = test['Close'].values
```

```python
# Standardize input features
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# Define the ANN model architecture neural network with three hidden layers
model_ann = Sequential([
    Dense(128, activation='relu', input_shape=(x_train.shape[1],)),         #Dense layers are fully connected layers
    Dropout(0.2),
    Dense(64, activation='relu'),         #activation function to be applied to the output
    Dropout(0.2),
    Dense(32, activation='relu'),       #Dropout is a regularization technique used to prevent overfitting 
    Dense(1)  #This neuron represents the output of the model.
])
```

```python
# Compile the model configure the learning process 
#algorithm used to update the weights of the neural network during training Adam
#momentum: accelerate convergence during the training process & RMSProp: preventing the learning rates from #diminishing rapidly
model_ann.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error') 
```

```python
# Train the model
# Verbose = 0  silent mode, no training progress is displayed during the training process You won't see any output printed to the console
# Verbose = 1 (Progress Bar):You will see an updated progress bar for each epoch, 
# Verbose = 2 (One Line per Epoch):Each line typically contains information about the current epoch number, the total number of epochs,
#and the average loss and any other metrics.
history = model_ann.fit(x_train_scaled, y_train, epochs=100, batch_size=32, validation_split=0.2, verbose=0)

```

```python
# Plot training and validation loss
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# Predictions on the test set
y_pred = model_ann.predict(x_test_scaled)

# Calculate evaluation metrics
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
medae = median_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print evaluation metrics
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)  #difference between the predicted values and the actual values
print("Median Absolute Error:", medae)#For each data point, calculate the absolute difference between the predicted value and the actual value then Calculate the median of all the absolute differences
print("R2 score:", r2)
```

```python
# Evaluate the model on test set how well it performs on unseen data.
loss = model_ann.evaluate(x_test_scaled, y_test)
print(f'Test Loss: {loss}')
```

```python
# Function to predict stock price
def predict_stock_price(open_price, high_price, low_price, volume):
    input_data = np.array([open_price, high_price, low_price, volume]).reshape(1, -1)
    input_data_scaled = scaler.transform(input_data)
    predicted_price = model_ann.predict(input_data_scaled)
    return predicted_price[0]
```
